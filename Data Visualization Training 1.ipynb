```python
usa = populasi[populasi.country == 'United States']
china = populasi[populasi.country == 'China']
indo = populasi[populasi.country == 'Indonesia']
```

```python
indo
```

```python
plt.plot(usa.year, usa.population)
plt.show()
```

```python
plt.plot(usa.year, usa.population / 10**6)   #pangkat
plt.show()
```

```python
plt.plot(usa.year, usa.population / 10**6)
plt.plot(china.year,  china.population / 10**6)
plt.plot(indo.year , indo.population / 10**6)
plt.legend(['USA', 'China', 'Indonesia'])
plt.ylabel('Population')
plt.xlabel('Year')

plt.show()
```

```python
indo_sum = indo.population / indo.population.iloc[0] * 100      #dalam persen %
```
